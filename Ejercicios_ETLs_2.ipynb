```
#13.1. Vamos a recordar a usar map.
"""
Repaso
Utilizaremos la función map cuando queremos aplicar una funcion a todos los elementos de una columna.

La función Map() toma como argumento otra función, que contiene el código que queremos aplicar a cada elemento de la lista o columna.
"""
#Planteamos el siguiente ejercicio:
#-Si un deportista tiene dos medallas o mas, le ponemos la etiquieta de Hall of Fame
#-Si un deportista tiene una medalla, le ponemos la etiquieta de Over-Average
#-Si un deportista tiene cero medallas, le ponemos la etiquieta de Average
#Para ellos hacemos una función que en función del valor de entrada le pongamos un valor u otro, llamamos a la funcion, ganador_juegos
def ganador_juegos(num):
  if (num == 2):
    return "Hall of fame"
  elif (num == 1):
    return "Over-Average"
  elif (num == 0):
    return "Average"
```

```
#13.2. Ahora creamos una columna nueva llamada Tipo_Deportista y utilizando MAP sobre la columna Numero_Medallas_Olimpicas
deportistas["Tipo_Deportista"] = deportistas.Numero_medallas_olimpicas.map(ganador_juegos)
deportistas
```

```
#14. Para finalizar vamos a arreglar el salario y convertilo a positivo. Hagamos esto en una linea de forma limpia y directa
deportistas.salarios = deportistas.salarios.map(lambda x: x if x > 0 else -x)
deportistas
```

```
#1. Creamos un dataframe nuevo, el cual se va a llamar productos y vamos a tener los siguientes valores. (Columna : Valores)
"""
Producto: Ipad Mini, Iphone 12, Samsung S8
Pais_Origen: Francia, EEUU, España
Fecha_Entrada_Stock: "2020-03-12", "2020-12-11", "2018-03-26"
Fecha_Venta: "2021-01-03", "2020-12-25", np.nan
Precio: 449.50, 846.75, 242.24
"""
productos = pd.DataFrame({
    "Producto":["Ipad Mini","Iphone 12","Samsung S20","Xiaomi Redmi Note 9"],
    "Pais_Origen": ["Francia", "EEUU", "España","España"],
    "Fecha_Entrada_Stock": ["2021-03-12", "2020-12-11", "2018-03-26", "2021-05-26"],
    "Fecha_Venta": ["2021-04-03 13:00:43", "2020-12-25 10:32:11", np.nan,"2021-05-28"],
    "Precio": [449.50, 846.75, 242.24,139.44]
})
productos
```

```
#5. Convertimos el formato de objeto por el de fecha utitlizando el método to_datetime, solo le tenemos que pasar como parámetro la columna a modificar.
productos.Fecha_Venta = pd.to_datetime(productos.Fecha_Venta)
productos
```

```
#6.1 Ahora cambiamos el formato de la otra columna (Fecha_Entrada_Stock) usando astype, convirtiendo el campo a datetime64[ns]
productos.Fecha_Entrada_Stock = productos.Fecha_Entrada_Stock.astype("datetime64[ns]")
productos
```

```
productos.Fecha_Venta = productos.Fecha_Venta.fillna(pd.Timestamp("2023-11-20"))
productos
```

```
#6.2 Ahora vamos a agregar los datos por mes, para ello vamos a utilizar resample, resample necesita tener un index con formato de fecha o no funciona.
# Antes de asignar el nuevo indice vamos a crear una nueva columna llamada Fecha_Venta_Bis para luego hacer restas
productos["Fecha_Venta_Bis"] = productos['Fecha_Venta']

# Por eso vamos a asignar el campo fecha_venta como index
productos.set_index("Fecha_Venta_Bis", inplace = True)
productos.resample("Y").sum() #tiene que ser sobre un indice que sea de campo fecha
```

```
#6.3 Ahora vamos a agrupar por año
```

```
#7. Ahora vamos a ver un formato de fecha distinto, se conoce como EPOCH o UNIX
"""
El tiempo Epoch es una magnitud que proporciona a la fecha del ordenador un valor numérico expresado en segundos, así como un punto de partida desde el que empezar a contar.
De esta manera, los PCs pueden interpretar las fechas como magnitudes legibles y calculables.

Como decíamos, este tiempo es un punto de partida utilizado para calcular la fecha, pero cada sistema operativo lo ha establecido de una manera diferente.
Mientras que en el mundo real el tiempo Epoch se establece a las 00:00:00 UTC (donde UTC es el tiempo universal coordinado), en los sistemas operativos se establece de la siguiente manera:
- MacOS (Apple): comienza el 1 de enero de 1904.
- Microsoft Windows: comienza el 1 de enero de 1601.
- UNIX y distribuciones Linux: comienza el 1 de enero de 1970. Esta fecha también se conoce como «UNIX time».
Como podéis ver, el tiempo Epoch es por lo tanto un valor arbitrario que se ha establecido según han querido en cada caso, no es algo estándar o universal sino simplemente una manera de hacer que el ordenador interprete la fecha y la hora.
En el caso del UNIX time, fue establecido por Dennis Ritchie y Ken Thompson (los creadores de UNIX) en la década de los 60.

Hora normal 	Segundos
1 minuto	60 segundos
1 hora	3600 segundos
1 día	86400 segundos
1 semana	604800 segundos
1 mes (30.44 días) 	2629743 segundos
1 año (365.24 días) 	 31556926 segundos
productos.Fecha_Venta
"""
#Por eso vamos a añadir una columna con tiempo epoch y le vamos a cambiar el formato en el siguiente paso
```

```
#8. Ahora vamos a convertir dicho valor epoch a datetime
# ¿Cómo? Igual que en el ejercicio 6 pero añadiendo un parámetro conocido como unit
# La unidad del arg (D, s, ms, us, ns) denota la unidad, que es un número entero o flotante.
#  Esto se basará en el origen. Por ejemplo, con unit = "ms" y origin = "unix" (el valor predeterminado), esto calcularía el número de milisegundos para el inicio de época de Unix.
```

```
#9.1. Vamos a ver como tratar las fechas, como sumar y restar
# Podemos restar o sumar dos fechas si son del mismo formato
productos["Fecha_Venta"] - productos["Fecha_Entrada_Stock"]
```

```
#10.1. Si le queremos dar formato específico usamos la libreria datetime ya que nos devuelve un df al cual podemos aplicarle un método específico de datetime.
(productos["Fecha_Venta"] - productos["Fecha_Entrada_Stock"]).dt.days
```

```
#11. Ahora vamos a añadir una nueva columna que nos diga el nº de dias en stock que ha estado el producto
productos["dias_en_stock"] = (productos["Fecha_Venta"] - productos["Fecha_Entrada_Stock"]).dt.days
productos
```

```
# DATETIME: El objeto datetime es un único objeto que contiene toda la información de un objeto date y un objeto time .
#12.1. Vemos como obtener la fecha de hoy, y el día, mes, año
pd.Timestamp.now()
```

```
productos["dias_desde_venta"] = (pd.Timestamp.now() - productos["Fecha_Venta"]).dt.days
productos
```

```
#12.2. Tenemos dos maneras de hacerlo restandole esa fecha a la columna de fecha_venta y creando una nueva columna
```

```
#12.3. Vamos a rellenar el NaN de la columna de dias en stock con el promedio de dias que estuvieron en stock
productos.fillna(productos.dias_en_stock.mean())
productos
```

```
#12.4. Vamos a hacer lo mismo para el resto de columnas numericas
```

```
#13.1. Otra alternativa, aplicamos un map para calcular los dias que han pasado desde que se vendieron hasta hoy.
# Creamos una función que simplemente reste la fecha de hoy menos la que se pasa por parámetro.
# Llamamos a la función resta_fecha
def resta_fecha (date):
  return (pd.Timestamp.now() - date).days
```

```
#13.2. Por ultimo creamos una columna nueva llamada dias_dedse_venta_bis y aplicamos un map
productos["Fecha_Venta_bis"] = productos.Fecha_Venta.map(resta_fecha)
productos
```

# Sección 4. Visualización
Vamos a utilizar plot([texto del enlace](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.html)) tiene diferentes opciones, una de ella, elegir que tipo de gráficos queremos utilizar. Nosotros vamos a probar las siguientes:
* "[bar](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html)" para diagramas de barras verticales.
* "[hist](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.hist.html)" es para histogramas.
* "[line](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.line.html)" es para gráficos de líneas.
* "[pie](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.pie.html)" es para gráficos de tarta.

También tiene los siguientes:
* "[area](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.area.html)" es para diagramas de área.
* "[barh](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.barh.html)" para diagramas de barras horizontales.
* "[box](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.box.html)" diagramas de caja.
* "[scatter](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html)" es para diagrama de dispersión.

Para ello vamos a utilizar el df que hemos usado en las secciones anteriores, **Deportistas**.

```
#1. Utilizamos el df de deportistas de la primera seccion, deportistas.
# Por defecto te coge un gráfico de lineas
deportistas.plot()
```

```
#2. Le podemos definir exactamente que valores queremos en los dos ejes (x,y)
deportistas.plot(x = "Nombre", y = "Numero_medallas_olimpicas")
```

```
#3. También podemos usar un histograma cambiando el tipo de gráfico
deportistas["Numero_medallas_olimpicas"].plot(kind = "hist")
```

```
#4. En el caso de usar un gráfico de tarta, primero contamos los diferentes tipos de valores (con value_counts) y luego usamos el tipo PIE
deportistas["Numero_medallas_olimpicas"].value_counts().plot(kind = "pie")
```

```
#5. Por último si queremos pintar un gráfico de barras, le definimos el eje X,Y y vemos una manera distinta de llamar a un gráfico distinto.
deportistas.plot.bar(x = "Nombre", y = "Numero_medallas_olimpicas")
```
