Es decir, el usuario me dice que quiere lanzar 100 veces y quier quiero contar cuantas veces ha salido el numero que el quiera, por ejemplo el 3. EL resultado en este caso el resultado final seria "HAS LANZADO 100 VECES Y HAN SALIDO 30 VECES EL NUMERO 3". sar la funcion random.randit(1,6)

CHALLANGE: Vamos a crear una funcion que pida al usario cuantas veces quiere lanzar y tengo que devolver el numero de veces que ha salido cada cara del dado

```python
#Ejercicio 1

def palabra():
  p=input("Introduce una palabra:")
  return(p)

def palabra_reves():
  p1=palabra() #guardo la palabra en la variable p1
  n=len(p1) #voy a guardar en n la longitud de mi palabra
  p_resultado=''
  for i in range(-1,-n-1,-1):
    p_resultado = p_resultado + p1[i]
  print("La palabra al revés es: ", p_resultado)
palabra_reves()
```

```python
#Ejercicio 3 (Gustavo clase)
def numero_inputs():
  n1= int(input("Introduce un numero que quieras contar de la cara de un dado:"))
  n2= int(input("Introduce cuantas veces quieres lanzar el dado:"))
  return([n1, n2])

def lanzamiento_dado():
  n_input= numero_inputs() # como se que he definido la funcion numero_inputs que devuelva una lista, por lo tanto  la variable n_input sera una lista con 2 valores
  n_contador=0 #defino la variable donde voy a almacenar el numero de veces que va a salir en nº que ha dicho el usuario
  for i in range(0,n_input[1]):
    lado_dado=random.randint(1,6) #simulo mi lanzmiento
    if lado_dado==n_input[0]: #evaluar si en numero del lanzamiento es igual al numero que quiero contar el usuario
      n_contador=n_contador+1
    else:
       n_contador=n_contador
  return(n_contador)
```

```python
# Ejercicio 3 (CHAT GPT)

def lanzar_dado_y_contar():
    veces_lanzar = int(input("¿Cuántas veces quieres lanzar el dado? "))

    if veces_lanzar <= 0:
        print("El número de lanzamientos debe ser mayor que 0.")
        return

    resultados = [random.randint(1, 6) for _ in range(veces_lanzar)]

    conteo_caras = [0, 0, 0, 0, 0, 0]
    for resultado in resultados:
        conteo_caras[resultado - 1] += 1

    for cara, cantidad in enumerate(conteo_caras):
        print(f"Cara {cara + 1} ha salido {cantidad} veces.")

lanzar_dado_y_contar()
```

```python
#Ejercicio 3 (Gustavo clase)

def conteo_caras():
  n_lanzamientos = int(input("¿Cuántas veces quieres lanzar el dado?"))
  contador_1=0
  contador_2=0
  contador_3=0
  contador_4=0
  contador_5=0
  contador_6=0
  for i in range(0,n_lanzamientos):
    lado_dado= random.randint(1,6)
    if lado_dado==1:
        contador_1=contador_1+1
    elif lado_dado==2:
        contador_2=contador_2+1
    elif lado_dado==3:
        contador_3=contador_3+1
    elif lado_dado==4:
        contador_4=contador_4+1
    elif lado_dado==5:
        contador_5=contador_5+1
    else:
        contador_6=contador_6+1
  return([contador_1,contador_2,contador_3,contador_4,contador_5,contador_6])

conteo_caras()
```

```python
fig=plt.figure(figsize=(10, 10))
fig.suptitle('GRAFICO DE IRIS', fontsize=20)

# Gráfico 1: Scatterplot de longitud del sépalo vs. ancho del pétalo (plt.scatter)
plt.subplot(2, 2, 1)
plt.scatter(df_iris["sepal.length"], df_iris["petal.width"])

# Gráfico 2: Gráfico de barras de variedad vs. longitud del sépalo
plt.subplot(2, 2, 2)
plt.bar(df_iris["variety"],df_iris["sepal.length"])

# Gráfico 3: Histograma de la longitud del pétalo
plt.subplot(2, 2, 3)
plt.hist(df_iris["petal.length"])

# Gráfico 4: Gráfico de líneas de la anchura del pétalo
plt.subplot(2, 2, 4)
plt.plot(df_iris["petal.width"])

# Ajusta el espacio entre los subplots para una mejor presentación
plt.tight_layout()

# Muestra la matriz de gráficos
plt.show()
```
